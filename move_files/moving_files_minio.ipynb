```python
from minio import Minio
from minio.error import S3Error
from minio.commonconfig import CopySource
import streamlit as st

# --- Load secrets ---
key = st.secrets["MINIO_KEY"]
secret = st.secrets["MINIO_SECRET"]

# --- Connect to MinIO ---
client = Minio(
    "minio.carlboettiger.info",
    access_key=key,
    secret_key=secret,
    secure=True  # change to False if you're not using HTTPS
)

# --- Define bucket and folder paths ---
BUCKET = "public-biodiversity"
SOURCE_PREFIX = "pad-us-4_1/"
DEST_PREFIX = "pad-us-4_1/raw/"

# --- List all objects under the source prefix ---
objects = client.list_objects(BUCKET, prefix=SOURCE_PREFIX, recursive=True)

for obj in objects:
    if obj.object_name.startswith(DEST_PREFIX):
        # skip files already in the destination
        continue

    # Build the new object name under the destination prefix
    new_object_name = obj.object_name.replace(SOURCE_PREFIX, DEST_PREFIX, 1)

    try:
        # Copy the object to the new location
        source = CopySource(BUCKET, obj.object_name)
        client.copy_object(BUCKET, new_object_name, source)
        st.write(f"Copied: {obj.object_name} â†’ {new_object_name}")

        # Remove the original object (to "move" instead of just copy)
        client.remove_object(BUCKET, obj.object_name)
        st.write(f"Deleted original: {obj.object_name}")

    except S3Error as e:
        st.write(f"Error moving {obj.object_name}: {e}")
```
